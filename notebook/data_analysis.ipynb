```python
def parse_dpd(hist):

    if pd.isna(hist) or hist == "":
        return []

    return [
        int(hist[i:i+3])
        for i in range(0, len(hist), 3)
    ]
```

```python
def recent12_max(x):
    if len(x) == 0:
        return 0
    return max(x[-12:])

def recent6_max(x):
    if len(x) == 0:
        return 0
    return max(x[-6:])

def recent3_max(x):
    if len(x) == 0:
        return 0
    return max(x[-3:])

def weighted_dpd(x):
    if len(x) == 0:
        return 0

    weights = np.arange(1, len(x)+1)

    return np.average(x,weights=weights)

def delinquency_ratio(x):
    if len(x) == 0:
        return 0

    return (sum(v > 0 for v in x)/ len(x))

def longest_streak(x):

    best = 0
    cur = 0

    for v in x:

        if v > 0:
            cur += 1
            best = max(best, cur)
        else:
            cur = 0

    return best

def dpd_volatility(x):

    if len(x) <= 1:
        return 0

    return np.std(x)

def recovery_score(x):

    if len(x) < 2:
        return 0

    return x[0] - x[-1]
```
